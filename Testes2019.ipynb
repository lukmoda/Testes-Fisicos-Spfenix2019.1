# Relatório Testes Físicos Fênix 2019.1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Testes2019.csv', header=1).set_index('Jogador')
df
```

# Tempos (Médio, Máximo, Mínimo)

\* Lembrar que o Broad Jump não é "Tempo"

```python
cols = [colum for colum in df.columns[0:4]]

d = {'Tempo Médio': [np.mean(df[colum]) for colum in df.columns[0:4]], 
     'Menor Tempo': [np.min(df[colum]) for colum in df.columns[0:4]],
     'Maior Tempo': [np.max(df[colum]) for colum in df.columns[0:4]]
    }

np.round(pd.DataFrame(data=d, index=cols), 2)
```

# 40 yard Dash

```python
df['40 yd dash'].sort_values().plot.bar(figsize=(16,8), color='b')
plt.xlabel('Jogador')
plt.ylabel('Tempo (s)')
plt.title('40 Yard Dash')
plt.show()
```

# 3 Cone Drill

```python
df['3 cone drill'].sort_values().plot.bar(figsize=(16,8), color='g')
plt.xlabel('Jogador')
plt.ylabel('Tempo (s)')
plt.title('3 Cone Drill')
plt.show()
```

# 5 10 5

```python
df['5 10 5'].sort_values().plot.bar(figsize=(16,8), color='r')
plt.xlabel('Jogador')
plt.ylabel('Tempo (s)')
plt.title('5 10 5')
plt.show()
```

# Broad Jump

```python
df['Broad Jump'].sort_values(ascending=False).plot.bar(figsize=(16,8), color='brown')
plt.xlabel('Jogador')
plt.ylabel('Distância')
plt.title('Broad Jump')
plt.show()
```

# Tempo Acumulado

```python
df['Tempo Acumulado'] = df['40 yd dash'] + df['3 cone drill'] + df['5 10 5']

df['Tempo Acumulado'].sort_values().plot.bar(figsize=(16,8), color='cyan')
plt.xlabel('Jogador')
plt.ylabel('Tempo Acumulado (s)')
plt.title('Tempo Acumulado')
plt.show()
```
